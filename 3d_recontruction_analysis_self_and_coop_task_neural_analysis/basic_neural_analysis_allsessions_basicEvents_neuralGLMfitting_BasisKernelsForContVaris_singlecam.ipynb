```python
def cluster_based_correction_with_timing(real_mean, shuffled_coefs, alpha_form, alpha_sig, time_axis=None):
    n_boot, n_vars, n_basis = shuffled_coefs.shape
    cluster_significance = np.zeros(n_vars, dtype=bool)
    cluster_timing = ['None'] * n_vars

    for var in range(n_vars):
        real_coef = real_mean[var, :]
        shuf_coef = shuffled_coefs[:, var, :]

        # Empirical p-values
        p_vals = (np.sum(np.abs(shuf_coef) >= np.abs(real_coef), axis=0) + 1) / (n_boot + 1)
        sig_mask = p_vals < alpha_form

        # Cluster label
        labeled_array, n_clusters = label(sig_mask)
        real_cluster_sizes = [
            np.sum(labeled_array == cluster_idx + 1)
            for cluster_idx in range(n_clusters)
        ]
        max_real_cluster = np.max(real_cluster_sizes) if real_cluster_sizes else 0

        # Compute max cluster size in each shuffled iteration
        shuf_max_clusters = []
        for b in range(n_boot):
            others = np.delete(shuf_coef, b, axis=0)
            p_vals_shuf = np.mean(np.abs(others) >= np.abs(shuf_coef[b, :]), axis=0)
            sig_shuf = p_vals_shuf < alpha_form
            lbl_shuf, n_lbl = label(sig_shuf)
            max_cluster = max([np.sum(lbl_shuf == i + 1) for i in range(n_lbl)], default=0)
            shuf_max_clusters.append(max_cluster)

        cluster_thresh = np.percentile(shuf_max_clusters, 100 * (1 - alpha_sig))
        cluster_significance[var] = max_real_cluster > cluster_thresh

        # Dominant timing classification
        if cluster_significance[var] and time_axis is not None:
            sig_times = time_axis[sig_mask]
            n_before = np.sum(sig_times < 0)
            n_after = np.sum(sig_times > 0)

            if n_before > n_after:
                cluster_timing[var] = 'Reactive'
            elif n_after > n_before:
                cluster_timing[var] = 'Predictive'
            else:
                cluster_timing[var] = 'Both' if (n_before > 0) else 'None'

    return cluster_significance, cluster_timing
```

```python
# check the orthogonality among the three behavioral vectors
def check_orthogonality(*vectors, tol=1e-6):
    n = len(vectors)
    for i in range(n):
        for j in range(i + 1, n):
            dot = np.dot(vectors[i], vectors[j])
            print(f"Dot product between vector {i} and {j}: {dot:.6f}")
            if abs(dot) > tol:
                print("⚠️ Not orthogonal!")
            else:
                print("✅ Orthogonal")
                
#  Gram-Schmidt Orthogonalization
def gram_schmidt(vectors):
    """Orthogonalize a list of vectors using the Gram-Schmidt process."""
    orthogonal_vectors = []
    for v in vectors:
        # Subtract projection onto all previous orthogonal vectors
        for u in orthogonal_vectors:
            v = v - np.dot(v, u) * u
        # Normalize the vector
        norm = np.linalg.norm(v)
        if norm > 1e-10:
            orthogonal_vectors.append(v / norm)
    return orthogonal_vectors
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# --- 1. Define Parameters ---

# Total number of time points for the simulation
time_points = 1200
# The standard deviation (sigma) for the Gaussian kernel.
# This value controls the amount of smoothing.
# Try changing it to see how it affects the output!
sigma = 15.0

# --- 2. Generate Example Binary Data ---

# Create a time axis from 0 to time_points-1
time = np.arange(time_points)

# Create three empty binary arrays (all zeros)
pull_series = np.zeros(time_points)
gaze_series = np.zeros(time_points)
juice_series = np.zeros(time_points)

# Define the time points where events (spikes) occur for each series
pull_event_times = [200, 600, 1000]
gaze_event_times = [300, 700, 750, 950, 965, 980, 990]
juice_event_times = [400, 800, 1100]

# Set the values at the event times to 1.0 in our binary arrays
pull_series[pull_event_times] = 1.0
gaze_series[gaze_event_times] = 1.0
juice_series[juice_event_times] = 1.0

# --- 3. Apply Gaussian Smoothing ---

# Use scipy's gaussian_filter1d to efficiently smooth the data.
# This function applies a Gaussian filter along the array.
pull_smoothed = gaussian_filter1d(pull_series, sigma=sigma)
gaze_smoothed = gaussian_filter1d(gaze_series, sigma=sigma)
juice_smoothed = gaussian_filter1d(juice_series, sigma=sigma)

# --- 4. Plot the Results ---

# Create a figure with 3 subplots, arranged vertically.
# `sharex=True` ensures all plots share the same time axis.
fig, axes = plt.subplots(3, 1, figsize=(10, 6), sharex=True)
fig.suptitle('Binary Time Series and Gaussian Smoothing', fontsize=16)

# Plotting settings for each series
plot_configs = [
    {
        'ax': axes[0],
        'title': 'Pull Axis Projection',
        'binary_series': pull_series,
        'smoothed_series': pull_smoothed,
        'binary_color': 'black',
        'linestyle': 'solid'
    },
    {
        'ax': axes[1],
        'title': 'Gaze Axis Projection',
        'binary_series': gaze_series,
        'smoothed_series': gaze_smoothed,
        'binary_color': '#8B4513', # SaddleBrown
        'linestyle': 'solid'
    },
    {
        'ax': axes[2],
        'title': 'Juice Axis Projection',
        'binary_series': juice_series,
        'smoothed_series': juice_smoothed,
        'binary_color': 'black',
        'linestyle': 'dashed'
    }
]

# Loop through the configurations to create each plot
for config in plot_configs:
    ax = config['ax']
    binary_series = config['binary_series']
    
    # Find the indices where events occur to plot the vertical lines
    event_indices = np.where(binary_series > 0)[0]
    
    # Plot the original binary events as vertical lines
    ax.vlines(event_indices, 0, 1,
              color=config['binary_color'],
              linestyle=config['linestyle'],
              linewidth=2.5,
              label='Binary Events')
              
    # Plot the smoothed curve
    ax.plot(time, config['smoothed_series'],
            color='dodgerblue',
            linewidth=2,
            label=f'Smoothed (σ={sigma})')

    # --- Formatting for a clean look ---
    ax.set_ylabel(config['title'], fontsize=12, rotation=0, labelpad=80)
    ax.set_yticks([]) # Remove y-axis ticks
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_position('center')
    ax.set_ylim(0, np.max(gaze_smoothed) * 1.1) # Ensure all curves are visible

# Set a label for the shared x-axis
axes[-1].set_xlabel('Time', fontsize=12)

# Adjust layout to prevent labels from overlapping
plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust for suptitle
# plt.show()

savefig = 0
if savefig:
    figsavefolder = data_saved_folder+"fig_for_basic_neural_analysis_allsessions_basicEvents_neuralGLMfitting_BasisKernelsForContVaris_singlecam/"+\
                    cameraID+"/"

    if not os.path.exists(figsavefolder):
        os.makedirs(figsavefolder)

    fig.savefig(figsavefolder+'PullGazeJuice_binary_and_smoothed_example.pdf')


```

```python
from statsmodels.stats.multitest import multipletests

# from p value, defining significant encoding type considering correction
indices = np.where(np.isin(task_conditions, conditions_to_ana))[0]
dates_to_ana = np.array(dates_list)[indices]

# load targeted pvalues
ind_con_tgt = np.isin(neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_pvalues_summary['date'],dates_to_ana)
pvalue_tgt = neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_pvalues_summary[ind_con_tgt]

# load targeted filters
tempFilter_tgt = {
    date: neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_tempFilter_summary[date]
    for date in dates_to_ana
    if date in neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_tempFilter_summary}


# Create a unique identifier for each neuron to handle data from different date
pvalue_tgt['unique_neuron'] = pvalue_tgt['date'].astype(str) + '_' + pvalue_tgt['neuronID'].astype(str)

# --- Step 1: Isolate the p-value columns for correction ---
p_value_cols = [
    'var_pull_past', 'var_pull_future', 'var_gaze_past', 'var_gaze_future',
    'var_juice_past', 'var_juice_future', 'partner_PC1_past', 'partner_PC1_future'
]
raw_p_values_df = pvalue_tgt[p_value_cols]


# --- Step 2: Perform the Global FDR Correction ---

# Flatten the DataFrame into a single 1D array.
original_shape = raw_p_values_df.shape
all_p_values = raw_p_values_df.values.flatten()

# The correction function cannot handle NaNs, so we create a mask to track them.
nan_mask = np.isnan(all_p_values)
p_values_for_correction = all_p_values[~nan_mask]

# Apply the Benjamini-Hochberg FDR correction.
is_significant_flat, q_values_flat, _, _ = multipletests(
    p_values_for_correction, 
    alpha=0.05,  # Standard significance level for FDR
    # method='fdr_bh'
    method='bonferroni'
)

# Reshape the corrected q-values back into the original DataFrame shape.
q_values_full = np.full(all_p_values.shape, np.nan)
q_values_full[~nan_mask] = q_values_flat
q_values_reshaped = q_values_full.reshape(original_shape)

# Create a final DataFrame for the q-values with the original index and columns.
q_values_df = pd.DataFrame(q_values_reshaped, index=raw_p_values_df.index, columns=raw_p_values_df.columns)

# This is the fully corrected boolean DataFrame indicating significance (q < 0.05).
significant_results_df = q_values_df < 0.05
# significant_results_df = q_values_df < 0.1


# --- Step 3: Classify Encoding Type for ALL Variables ---

# Define a GENERALIZED function to classify based on corrected significance
def classify_encoding(row, base_variable_name):
    """
    Classifies encoding type for a given base variable name.
    Example: base_variable_name = 'var_pull'
    """
    sig_past = row[f'{base_variable_name}_past']
    sig_future = row[f'{base_variable_name}_future']
    
    if sig_past and sig_future:
        return 'Both'
    elif sig_past:
        return 'Reactive'
    elif sig_future:
        return 'Predictive'
    else:
        return 'None'

# Apply the function for each of the four behavioral variables
pull_encoding_type = significant_results_df.apply(lambda row: classify_encoding(row, 'var_pull'), axis=1)
gaze_encoding_type = significant_results_df.apply(lambda row: classify_encoding(row, 'var_gaze'), axis=1)
juice_encoding_type = significant_results_df.apply(lambda row: classify_encoding(row, 'var_juice'), axis=1)
partner_encoding_type = significant_results_df.apply(lambda row: classify_encoding(row, 'partner_PC1'), axis=1)


# --- Step 4: Create the Final, Expanded Summary DataFrame ---

summary_df = pd.DataFrame({
    'unique_neuron': pvalue_tgt['unique_neuron'],
    'pull_encoding': pull_encoding_type,
    'gaze_encoding': gaze_encoding_type,
    'juice_encoding': juice_encoding_type,
    'partner_PC1_encoding': partner_encoding_type
})
```

```python
# =========================================================
# Integrated cell (updated selection):
# (A) build corrected significance + encoding labels from your pvalue summary
# (B) select example neurons
#     - PULL: GLM category + peak-time constraint (Predictive peak<0, Reactive peak>0)
#     - GAZE: GLM p-value based (no peak-time constraint)
# (C) plot ALL behavioral events per selected neuron (mean±SEM across trials)
# =========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.multitest import multipletests

# ----------------------------
# Part A. Build corrected significance + encoding labels from p-values
# ----------------------------
indices = np.where(np.isin(task_conditions, conditions_to_ana))[0]
dates_to_ana = np.array(dates_list)[indices]

# load targeted pvalues
ind_con_tgt = np.isin(neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_pvalues_summary['date'], dates_to_ana)
pvalue_tgt = neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_pvalues_summary[ind_con_tgt].copy()

# load targeted filters (kept as you had; not used below, but leaving it here)
tempFilter_tgt = {
    date: neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_tempFilter_summary[date]
    for date in dates_to_ana
    if date in neuralGLM_mainAxes_partnerPC1_LOOmethod_kernels_tempFilter_summary
}

# Create a unique identifier for each neuron to handle data from different date
pvalue_tgt['unique_neuron'] = pvalue_tgt['date'].astype(str) + '_' + pvalue_tgt['neuronID'].astype(str)

# --- Step 1: Isolate the p-value columns for correction ---
p_value_cols = [
    'var_pull_past', 'var_pull_future', 'var_gaze_past', 'var_gaze_future',
    'var_juice_past', 'var_juice_future', 'partner_PC1_past', 'partner_PC1_future'
]
raw_p_values_df = pvalue_tgt[p_value_cols].copy()

# --- Step 2: Perform the Global correction ---
original_shape = raw_p_values_df.shape
all_p_values = raw_p_values_df.values.flatten()

nan_mask = np.isnan(all_p_values)
p_values_for_correction = all_p_values[~nan_mask]

# NOTE: keeping your method as-is
is_significant_flat, q_values_flat, _, _ = multipletests(
    p_values_for_correction,
    alpha=0.05,
    method='fdr_bh'
    # method='bonferroni'
)

q_values_full = np.full(all_p_values.shape, np.nan)
q_values_full[~nan_mask] = q_values_flat
q_values_reshaped = q_values_full.reshape(original_shape)

q_values_df = pd.DataFrame(q_values_reshaped, index=raw_p_values_df.index, columns=raw_p_values_df.columns)
significant_results_df = q_values_df < 0.05

# --- Step 3: Classify Encoding Type for ALL Variables ---
def classify_encoding(row_bool, base_variable_name):
    """
    Classifies encoding type for a given base variable name.
    Example: base_variable_name = 'var_pull'
    row_bool is a boolean Series for significance (corrected q < 0.05)
    """
    sig_past = bool(row_bool[f'{base_variable_name}_past'])
    sig_future = bool(row_bool[f'{base_variable_name}_future'])

    if sig_past and sig_future:
        return 'Both'
    elif sig_past:
        return 'Reactive'
    elif sig_future:
        return 'Predictive'
    else:
        return 'None'


# ----------------------------
# Part B. Build enc_df + select examples
# ----------------------------
def safe_neglog10(x):
    x = np.asarray(x, dtype=float)
    x = np.clip(x, 1e-300, 1.0)
    return -np.log10(x)

rows = []
for idx in q_values_df.index:
    sig_row = significant_results_df.loc[idx]
    out = {
        'idx': idx,
        'dates': str(pvalue_tgt.loc[idx, 'date']),
        'neuronID': int(pvalue_tgt.loc[idx, 'neuronID']),
        'unique_neuron': pvalue_tgt.loc[idx, 'unique_neuron'],
    }

    for bv in ['var_pull', 'var_gaze']:
        out[f'{bv}_encoding'] = classify_encoding(sig_row, bv)
        out[f'{bv}_q_past'] = float(q_values_df.loc[idx, f'{bv}_past'])
        out[f'{bv}_q_future'] = float(q_values_df.loc[idx, f'{bv}_future'])
        out[f'{bv}_sig_past'] = bool(sig_row[f'{bv}_past'])
        out[f'{bv}_sig_future'] = bool(sig_row[f'{bv}_future'])

    rows.append(out)

enc_df = pd.DataFrame(rows)

# strength scores for ranking (used for gaze selection; pull selection will use peak timing)
enc_df['gaze_reactive_strength']   = safe_neglog10(enc_df['var_gaze_q_past'])
enc_df['gaze_predictive_strength'] = safe_neglog10(enc_df['var_gaze_q_future'])
enc_df['gaze_both_strength']       = safe_neglog10(np.minimum(enc_df['var_gaze_q_past'], enc_df['var_gaze_q_future']))

enc_df['pull_reactive_strength']   = safe_neglog10(enc_df['var_pull_q_past'])
enc_df['pull_predictive_strength'] = safe_neglog10(enc_df['var_pull_q_future'])
enc_df['pull_both_strength']       = safe_neglog10(np.minimum(enc_df['var_pull_q_past'], enc_df['var_pull_q_future']))


# ---- Pull selection: GLM category + peak time constraint ----
def get_event_mean_trace(bhvevents_dict, dates, bhv_plot, neuronID_str, t_fallback=(-4.0, 4.0)):
    dneur = bhvevents_dict[dates][bhv_plot][neuronID_str]
    fr_all = np.asarray(dneur['FR_allevents'])

    # mean across trials (your typical: (ntime, ntrials))
    if fr_all.ndim == 2 and fr_all.shape[0] >= fr_all.shape[1]:
        mean = np.nanmean(fr_all, axis=1)
    elif fr_all.ndim == 2:
        mean = np.nanmean(fr_all, axis=0)
    else:
        mean = fr_all

    # time vector
    tvec = None
    for k in ['tvec', 'timevec', 'time_vec', 'ts', 't', 'time']:
        if k in dneur:
            tvec = np.asarray(dneur[k])
            break
    if tvec is None:
        tvec = np.linspace(t_fallback[0], t_fallback[1], len(mean))

    nmin = min(len(tvec), len(mean))
    return tvec[:nmin], mean[:nmin]

def peak_time_baseline_subtracted(tvec, mean, search_win=(-2, 2), baseline_win=(-4, -2)):
    tvec = np.asarray(tvec)
    mean = np.asarray(mean)

    bmask = (tvec >= baseline_win[0]) & (tvec <= baseline_win[1])
    if np.sum(bmask) > 0:
        mean0 = mean - np.nanmean(mean[bmask])
    else:
        mean0 = mean - np.nanmean(mean)

    smask = (tvec >= search_win[0]) & (tvec <= search_win[1])
    if np.sum(smask) < 3:
        return np.nan, np.nan

    y = mean0[smask]
    t = tvec[smask]
    m = np.isfinite(y) & np.isfinite(t)
    if np.sum(m) < 3:
        return np.nan, np.nan

    y = y[m]; t = t[m]
    k = int(np.nanargmax(y))
    return float(t[k]), float(y[k])

def add_pull_peak_metrics(df, bhvevents_dict, act_animal_to_ana, bhv_event='pull',
                          search_win=(-2, 2), baseline_win=(-4, -2)):
    bhv_plot = f"{act_animal_to_ana} {bhv_event}"
    df = df.copy()
    peak_t_list = []
    peak_amp_list = []

    for r in df.itertuples():
        try:
            tvec, mean = get_event_mean_trace(bhvevents_dict, r.dates, bhv_plot, str(int(r.neuronID)))
            pt, pa = peak_time_baseline_subtracted(tvec, mean, search_win=search_win, baseline_win=baseline_win)
        except Exception:
            pt, pa = np.nan, np.nan
        peak_t_list.append(pt)
        peak_amp_list.append(pa)

    df['pull_peak_t'] = peak_t_list
    df['pull_peak_amp'] = peak_amp_list
    return df

def select_pull_examples_glm_plus_peak(enc_df, bhvevents_dict, act_animal_to_ana,
                                      n_each=2,
                                      pre_peak_win=(-2, 0),
                                      post_peak_win=(0, 2),
                                      min_peak_amp=0.05,
                                      prefer_unique_dates=True):
    df = enc_df.copy()
    df = add_pull_peak_metrics(df, bhvevents_dict, act_animal_to_ana, bhv_event='pull')

    # keep usable peaks
    df = df[np.isfinite(df['pull_peak_t']) & np.isfinite(df['pull_peak_amp'])]
    df = df[df['pull_peak_amp'] >= min_peak_amp]

    # GLM category filters
    df_pred = df[(df['var_pull_encoding'] == 'Predictive') &
               (df['pull_peak_t'] >= pre_peak_win[0]) & (df['pull_peak_t'] <= pre_peak_win[1])].copy()
    # df_pred = df[(df['var_pull_encoding'] == 'Predictive')].copy()

    df_reac = df[(df['var_pull_encoding'] == 'Reactive') &
                 # (df['pull_peak_t'] >= post_peak_win[0]) & (df['pull_peak_t'] <= post_peak_win[1])].copy()
                 (df['pull_peak_t'] >= pre_peak_win[0]) & (df['pull_peak_t'] <= pre_peak_win[1])].copy()
    # df_reac = df[(df['var_pull_encoding'] == 'Reactive')].copy()
    
    # df_both = df[(df['var_pull_encoding'] == 'Both')].copy()
    df_both = df[(df['var_pull_encoding'] == 'Both') &
                 (df['pull_peak_t'] >= pre_peak_win[0]) & (df['pull_peak_t'] <= pre_peak_win[1])].copy()

    # rank within each by peak amplitude (visual clarity), then de-duplicate dates
    def finalize(d, label):
        d = d.sort_values('pull_peak_amp', ascending=False)
        if prefer_unique_dates:
            d = d.drop_duplicates('dates')
        d = d.head(n_each).copy()
        d['example_group'] = f'pull:{label}'
        return d

    # ex_pred = finalize(df_pred, 'Predictive(peak<0)')
    # ex_reac = finalize(df_reac, 'Reactive(peak>0)')
    ex_pred = finalize(df_pred, 'Predictive')
    ex_reac = finalize(df_reac, 'Reactive')
    ex_both = finalize(df_both, 'Both')

    examples_pull = pd.concat([ex_reac, ex_pred, ex_both], ignore_index=True)

    # if any category is short, fall back to GLM-only ranking for that category
    def fallback_glm_only(label, enc_value, score_col):
        dd = enc_df[enc_df['var_pull_encoding'] == enc_value].copy()
        dd = dd[np.isfinite(dd[score_col])].sort_values(score_col, ascending=False)
        if prefer_unique_dates:
            dd = dd.drop_duplicates('dates')
        dd = dd.head(n_each).copy()
        dd['example_group'] = f'pull:{label}(GLM-only)'
        dd['pull_peak_t'] = np.nan
        dd['pull_peak_amp'] = np.nan
        return dd

    if np.sum(examples_pull['example_group'].str.contains('Reactive')) < n_each:
        ex_reac_fb = fallback_glm_only('Reactive', 'Reactive', 'pull_reactive_strength')
        examples_pull = pd.concat([examples_pull[~examples_pull['example_group'].str.contains('Reactive')],
                                   ex_reac_fb, examples_pull[examples_pull['example_group'].str.contains('Reactive')]],
                                  ignore_index=True)

    if np.sum(examples_pull['example_group'].str.contains('Predictive')) < n_each:
        ex_pred_fb = fallback_glm_only('Predictive', 'Predictive', 'pull_predictive_strength')
        examples_pull = pd.concat([examples_pull[~examples_pull['example_group'].str.contains('Predictive')],
                                   ex_pred_fb, examples_pull[examples_pull['example_group'].str.contains('Predictive')]],
                                  ignore_index=True)

    if np.sum(examples_pull['example_group'].str.contains('Both')) < n_each:
        ex_both_fb = fallback_glm_only('Both', 'Both', 'pull_both_strength')
        examples_pull = pd.concat([examples_pull[~examples_pull['example_group'].str.contains('Both')],
                                   ex_both_fb, examples_pull[examples_pull['example_group'].str.contains('Both')]],
                                  ignore_index=True)

    # keep exactly 2+2+2 in order Reactive, Predictive, Both
    examples_pull = pd.concat([
        examples_pull[examples_pull['example_group'].str.contains('Reactive')].head(n_each),
        examples_pull[examples_pull['example_group'].str.contains('Predictive')].head(n_each),
        examples_pull[examples_pull['example_group'].str.contains('Both')].head(n_each),
    ], ignore_index=True)

    return examples_pull


# ---- Gaze selection: GLM p-value based (no peak constraint) ----
def select_top_neurons(df, cond, score_col, n=2, prefer_unique_dates=True):
    d = df[cond].copy()
    d = d[np.isfinite(d[score_col])]
    d = d.sort_values(score_col, ascending=False)
    if prefer_unique_dates:
        d = d.drop_duplicates('dates')
    return d.head(n)

def get_gaze_examples_glm_only(df, n_each=2, only=True, prefer_unique_dates=True):
    """
    Gaze examples: GLM-only ranking by corrected p-value strength (no peak timing constraint).
    only=True attempts to avoid pull-significant neurons; relaxes if needed.
    """
    enc_col = 'var_gaze_encoding'
    other_sig = (df['var_pull_sig_past'] | df['var_pull_sig_future'])

    reactive_cond   = (df[enc_col] == 'Reactive')
    predictive_cond = (df[enc_col] == 'Predictive')
    both_cond       = (df[enc_col] == 'Both')

    if only:
        reactive_cond   &= ~other_sig
        predictive_cond &= ~other_sig
        both_cond       &= ~other_sig

    ex_reac = select_top_neurons(df, reactive_cond,   'gaze_reactive_strength',   n_each, prefer_unique_dates)
    ex_pred = select_top_neurons(df, predictive_cond, 'gaze_predictive_strength', n_each, prefer_unique_dates)
    ex_both = select_top_neurons(df, both_cond,       'gaze_both_strength',       n_each, prefer_unique_dates)

    def ensure_n(ex, base_cond, score):
        if len(ex) >= n_each:
            return ex
        return select_top_neurons(df, base_cond, score, n_each, prefer_unique_dates)

    if only:
        ex_reac = ensure_n(ex_reac, (df[enc_col] == 'Reactive'),   'gaze_reactive_strength')
        ex_pred = ensure_n(ex_pred, (df[enc_col] == 'Predictive'), 'gaze_predictive_strength')
        ex_both = ensure_n(ex_both, (df[enc_col] == 'Both'),       'gaze_both_strength')

    ex_reac = ex_reac.assign(example_group='gaze:Reactive')
    ex_pred = ex_pred.assign(example_group='gaze:Predictive')
    ex_both = ex_both.assign(example_group='gaze:Both')

    return pd.concat([ex_reac, ex_pred, ex_both], ignore_index=True)


# ---- actually pick examples ----
examples_pull = select_pull_examples_glm_plus_peak(
    enc_df,
    bhvevents_aligned_FR_allevents_all_dates,
    act_animal_to_ana,
    n_each=2,
    pre_peak_win=(-2, 0),
    post_peak_win=(0, 2),
    min_peak_amp=0.05,
    prefer_unique_dates=True
)

examples_gaze = get_gaze_examples_glm_only(
    enc_df,
    n_each=2,
    only=True,
    prefer_unique_dates=True
)

print("=== Pull examples (GLM + peak timing) ===")
print(examples_pull[['dates','neuronID','example_group','var_pull_encoding','var_gaze_encoding',
                     'var_pull_q_past','var_pull_q_future','pull_peak_t','pull_peak_amp']])

print("\n=== Gaze examples (GLM-only) ===")
print(examples_gaze[['dates','neuronID','example_group','var_gaze_encoding','var_pull_encoding',
                     'var_gaze_q_past','var_gaze_q_future']])


# ----------------------------
# Part C. Plot ALL behavioral events per selected neuron
# ----------------------------
def get_trace_mean_sem(bhvevents_dict, dates, bhv_plot, neuronID_str, t_fallback=(-4.0, 4.0)):
    dneur = bhvevents_dict[dates][bhv_plot][neuronID_str]
    fr_all = np.asarray(dneur['FR_allevents'])

    # time vector
    tvec = None
    for k in ['tvec', 'timevec', 'time_vec', 'ts', 't', 'time']:
        if k in dneur:
            tvec = np.asarray(dneur[k])
            break
    if tvec is None:
        ntime = fr_all.shape[0] if fr_all.ndim >= 1 else len(fr_all)
        tvec = np.linspace(t_fallback[0], t_fallback[1], ntime)

    # mean ± SEM across trials (your data usually: (ntime, ntrials))
    if fr_all.ndim == 1:
        mean = fr_all
        sem = np.full_like(mean, np.nan, dtype=float)
    elif fr_all.ndim == 2:
        if fr_all.shape[0] >= fr_all.shape[1]:
            mean = np.nanmean(fr_all, axis=1)
            sem  = np.nanstd(fr_all, axis=1) / np.sqrt(np.sum(~np.isnan(fr_all), axis=1).clip(min=1))
        else:
            mean = np.nanmean(fr_all, axis=0)
            sem  = np.nanstd(fr_all, axis=0) / np.sqrt(np.sum(~np.isnan(fr_all), axis=0).clip(min=1))
    else:
        fr2 = fr_all.reshape(fr_all.shape[0], -1)
        mean = np.nanmean(fr2, axis=1)
        sem  = np.nanstd(fr2, axis=1) / np.sqrt(np.sum(~np.isnan(fr2), axis=1).clip(min=1))

    nmin = min(len(tvec), len(mean))

    # ---- fill small gaps for plotting ----
    def _interp_nan(y):
        y = np.asarray(y, float)
        x = np.arange(len(y))
        m = np.isfinite(y)
        if m.sum() < 2:
            return y
        y2 = y.copy()
        y2[~m] = np.interp(x[~m], x[m], y[m])
        return y2

    mean = _interp_nan(mean[:nmin])
    sem  = _interp_nan(sem[:nmin])

    return tvec[:nmin], mean, sem

def plot_examples_all_behaviors(examples_df, title, behaviors, act_animal_to_ana, bhvevents_dict,
                                figsize_per_col=3.8, figsize_per_row=2.6):

    examples_df = examples_df.copy()
    examples_df['neuronID_str'] = examples_df['neuronID'].astype(int).astype(str)

    n_neurons = len(examples_df)
    n_bhv = len(behaviors)

    fig, axes = plt.subplots(
        n_bhv, n_neurons,
        figsize=(figsize_per_col*n_neurons, figsize_per_row*n_bhv),
        sharex=True, # sharey='row'
    )

    if n_bhv == 1 and n_neurons == 1:
        axes = np.array([[axes]])
    elif n_bhv == 1:
        axes = axes[np.newaxis, :]
    elif n_neurons == 1:
        axes = axes[:, np.newaxis]

    fig.suptitle(title, fontsize=16, fontweight='bold')

    # column titles
    for col in range(n_neurons):
        r = examples_df.iloc[col]
        axes[0, col].set_title(
            f"{r['example_group']}\n{r['dates']} | n{r['neuronID']}",
            fontsize=11, fontweight='bold'
        )

    for i, bhv in enumerate(behaviors):
        bhv_plot = f"{act_animal_to_ana} {bhv}"

        for col in range(n_neurons):
            ax = axes[i, col]
            r = examples_df.iloc[col]
            dates = r['dates']
            nid = r['neuronID_str']

            try:
                tvec, mean, sem = get_trace_mean_sem(bhvevents_dict, dates, bhv_plot, nid)
                ax.plot(tvec, mean, lw=1.8)
                if np.any(np.isfinite(sem)):
                    ax.fill_between(tvec, mean-sem, mean+sem, alpha=0.25)
                ax.axvline(0, ls='--', lw=1)
            except KeyError:
                ax.text(0.5, 0.5, 'missing', transform=ax.transAxes,
                        ha='center', va='center')
                ax.axvline(0, ls='--', lw=1)

            if col == 0:
                ax.set_ylabel(bhv_plot, fontsize=10)
            if i == n_bhv - 1:
                ax.set_xlabel('Time (s)', fontsize=10)

            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            
            ax.set_ylim([-1,2.5])

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    return fig

# ---- run plots ----
# behaviors = ['pull', 'succpull', 'failpull', 'gaze', 'gazestart']
behaviors = ['pull', 'succpull', 'failpull']

fig = plot_examples_all_behaviors(
    examples_pull,
    # title="Example neurons: Pull (GLM category + peak timing constraint)",
    title="Example neurons: Pull (GLM category)",
    behaviors=behaviors,
    act_animal_to_ana=act_animal_to_ana,
    bhvevents_dict=bhvevents_aligned_FR_allevents_all_dates
)

savefig = 1
if savefig:
    figsavefolder = (
        data_saved_folder
        + "fig_for_basic_neural_analysis_allsessions_basicEvents_neuralGLMfitting_BasisKernelsForContVaris_singlecam/"
        + cameraID+"/"+savefile_sufix+"/" 
        + recordedanimals[0] + "_neuralGLM/"
    )

    if not os.path.exists(figsavefolder):
        os.makedirs(figsavefolder)

    fig.savefig(
        figsavefolder
        + recordedanimals[0]
        + "_singleUnitExamples_pull_GLMplusPeak_"
        + savefile_sufix
        + "_"
        + cond_toplot_type
        + ".pdf"
    )
    

########
fig = plot_examples_all_behaviors(
    examples_gaze,
    title="Example neurons: Gaze (GLM-only selection)",
    behaviors=behaviors,
    act_animal_to_ana=act_animal_to_ana,
    bhvevents_dict=bhvevents_aligned_FR_allevents_all_dates
)

if savefig:
    fig.savefig(
        figsavefolder
        + recordedanimals[0]
        + "_singleUnitExamples_gaze_GLMplusPeak_"
        + savefile_sufix
        + "_"
        + cond_toplot_type
        + ".pdf"
    )
    
    
    
    
    
```

```python
# =========================================================
# Partner-aligned population PSTHs, grouped by PARTNER_PC1 encoding type
# - Align neural activity to PARTNER events: pull / succpull / failpull / gaze / gazestart
# - Partner identity is inferred PER DATE from bhv-event dict keys (the animal name that is NOT act_animal_to_ana)
# - Neuron grouping uses: partner_PC1_encoding in summary_df (Reactive / Predictive / Both / None)
# - Robust parsing of unique_neuron when dates contain underscores (e.g., 20250430_MC)
# =========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pickle

# ----------------------------
# helper: infer partner name for a given date (based on dict keys)
# ----------------------------
def infer_partner_name_for_date(bhvevents_dict, date_str, actor_name):
    """
    bhvevents_dict[date_str] keys look like: "<animal> pull", "<animal> gaze", ...
    We pick the first-token animal name that is NOT actor_name.
    """
    if date_str not in bhvevents_dict:
        return None

    klist = list(bhvevents_dict[date_str].keys())
    animals = []
    for k in klist:
        if isinstance(k, str) and (' ' in k):
            animals.append(k.split(' ')[0])
    animals = list(pd.unique(animals))

    cand = [a for a in animals if a != actor_name]
    if len(cand) > 0:
        return cand[0]

    if len(animals) > 0:
        return animals[0]

    return None


# ----------------------------
# helper: robust mean trace from FR_allevents
# ----------------------------
def compute_fr_mean_timecourse(FR_allevents, expected_ntime=(240, 241)):
    fr = np.asarray(FR_allevents)

    if fr.ndim == 1:
        return fr

    if fr.ndim == 2:
        # typical: (ntime, ntrials)
        if fr.shape[0] in expected_ntime and fr.shape[1] not in expected_ntime:
            return np.nanmean(fr, axis=1)
        # sometimes: (ntrials, ntime)
        if fr.shape[1] in expected_ntime and fr.shape[0] not in expected_ntime:
            return np.nanmean(fr, axis=0)
        # fallback: assume longer dim is time
        if fr.shape[0] >= fr.shape[1]:
            return np.nanmean(fr, axis=1)
        else:
            return np.nanmean(fr, axis=0)

    # higher dims: flatten non-time dims conservatively
    fr2 = fr.reshape(fr.shape[0], -1)
    if fr2.shape[0] in expected_ntime:
        return np.nanmean(fr2, axis=1)

    fr3 = fr.reshape(-1, fr.shape[-1])
    return np.nanmean(fr3, axis=0)


# =========================================================
# main
# =========================================================
if 1:

    # --- 1. load aligned traces dict ---
    do_OFC = 1
    do_DLPFC  = 0
    if do_OFC:
        savefile_sufix_2 = '_OFCs'
    elif do_DLPFC:
        savefile_sufix_2 = '_DLPFCs'
    else:
        savefile_sufix_2 = ''

    data_saved_folder_2 = '/gpfs/radev/pi/nandy/jadi_gibbs_data/VideoTracker_SocialInter/3d_recontruction_analysis_self_and_coop_task_data_saved/'
    data_saved_subfolder_2 = (data_saved_folder_2
                              + 'data_saved_singlecam_wholebody' + savefile_sufix_2 + '/'
                              + cameraID + '/'
                              + animal1_fixedorders[0] + animal2_fixedorders[0] + '/')

    with open(data_saved_subfolder_2
              + '/bhvevents_aligned_FR_allevents_all_dates_'
              + animal1_fixedorders[0] + animal2_fixedorders[0] + '.pkl', 'rb') as f:
        bhvevents_aligned_FR_allevents_all_dates = pickle.load(f)

    # --- 2. build neuron table: dates + neuronID + partner_PC1_encoding ---
    df_neurons = summary_df.copy()

    if 'unique_neuron' not in df_neurons.columns:
        raise ValueError("summary_df must contain 'unique_neuron'.")

    # NOTE: adjust this if your actual column name differs
    
    partner_enc_col = 'partner_PC1_encoding'  #'partner_PC1_encoding'
    
    if partner_enc_col not in df_neurons.columns:
        raise ValueError(f"summary_df must contain '{partner_enc_col}' (Reactive/Predictive/Both/None).")

    # robust parse: split on LAST underscore so dates can contain underscores (e.g., 20250430_MC)
    if ('dates' not in df_neurons.columns) or ('neuronID' not in df_neurons.columns):
        tmp = df_neurons['unique_neuron'].astype(str).str.rsplit('_', n=1, expand=True)
        df_neurons['dates'] = tmp[0]
        df_neurons['neuronID'] = tmp[1].astype(int)

    # limit to the dates/conditions you are analyzing (consistent with your pipeline)
    indices = np.where(np.isin(task_conditions, conditions_to_ana))[0]
    dates_to_ana = np.array(dates_list)[indices]
    df_neurons = df_neurons[np.isin(df_neurons['dates'].astype(str), dates_to_ana)].copy()

    # --- 3. Plot settings ---
    categories = ['Reactive', 'Predictive', 'Both', 'None']
    ncats = len(categories)

    partner_behaviors = ['pull', 'succpull', 'failpull', 'gaze', 'gazestart']
    nbhvs = len(partner_behaviors)

    fig, axes = plt.subplots(nbhvs, ncats, figsize=(5 * ncats, 5 * nbhvs),
                             sharex=True, sharey='row')
    fig.suptitle('Mean PARTNER-event-aligned Traces by partner_PC1 Encoding Type',
                 fontsize=20, fontweight='bold')

    # --- 4. Loop over partner behaviors x encoding categories ---
    for i, bhv_short in enumerate(partner_behaviors):
        for j, cat_name in enumerate(categories):
            ax = axes[i, j]

            # subset neurons by partner_PC1 encoding
            df_subset = df_neurons[df_neurons[partner_enc_col] == cat_name].copy()
            num_neurons = df_subset.shape[0]

            neuron_mean_traces = []
            tvec_panel = None

            for ineuron in range(num_neurons):
                date_ineuron = str(df_subset.iloc[ineuron]['dates'])
                neuronID_ineuron = str(int(df_subset.iloc[ineuron]['neuronID']))

                # infer partner name per date
                partner_name = infer_partner_name_for_date(
                    bhvevents_aligned_FR_allevents_all_dates,
                    date_ineuron,
                    act_animal_to_ana
                )
                if partner_name is None:
                    continue

                bhv_plot = f"{partner_name} {bhv_short}"

                # pull trace dict
                try:
                    dtmp = bhvevents_aligned_FR_allevents_all_dates[date_ineuron][bhv_plot][neuronID_ineuron]
                    FR_allevents = dtmp['FR_allevents']
                except Exception:
                    continue

                # time vector
                tvec = None
                for k in ['tvec', 'timevec', 'time_vec', 'ts', 't', 'time']:
                    if k in dtmp:
                        tvec = np.asarray(dtmp[k])
                        break

                fr_mean = compute_fr_mean_timecourse(FR_allevents)

                # fallback time vector
                if tvec is None:
                    tvec = np.linspace(-4.0, 4.0, len(fr_mean))

                # length match
                nmin = min(len(tvec), len(fr_mean))
                tvec_use = np.asarray(tvec[:nmin])
                fr_mean = np.asarray(fr_mean[:nmin])

                if tvec_panel is None:
                    tvec_panel = tvec_use

                neuron_mean_traces.append(fr_mean)

            # --- plot mean +/- SEM across neurons ---
            if len(neuron_mean_traces) == 0:
                ax.text(0.5, 0.5, 'n=0', transform=ax.transAxes,
                        ha='center', va='center', fontsize=14)
                ax.axvline(0, linestyle='--', linewidth=1)
                ax.set_title(f'{cat_name} (n=0)', fontsize=14, fontweight='bold')
            else:
                lens = [len(x) for x in neuron_mean_traces]
                min_len = int(np.min(lens))
                traces_common = [np.asarray(x[:min_len]) for x in neuron_mean_traces]
                tplot = np.asarray(tvec_panel[:min_len])

                X = np.vstack(traces_common)  # (nneurons, ntime)
                mean_trace = np.nanmean(X, axis=0)
                sem_trace = np.nanstd(X, axis=0) / np.sqrt(np.sum(~np.isnan(X), axis=0).clip(min=1))

                ax.plot(tplot, mean_trace, linewidth=2)
                ax.fill_between(tplot, mean_trace - sem_trace, mean_trace + sem_trace, alpha=0.25)
                ax.axvline(0, linestyle='--', linewidth=1)

                ax.set_title(f'{cat_name} (n={X.shape[0]})', fontsize=14, fontweight='bold')

            if j == 0:
                ax.set_ylabel(f"PARTNER {bhv_short}", fontsize=14)
            if i == (nbhvs - 1):
                ax.set_xlabel('Time (aligned)', fontsize=14)

            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)

    plt.tight_layout(rect=[0, 0, 1, 0.97])

    # --- save (same style as you do elsewhere) ---
    savefig = 1
    if savefig:
        figsavefolder = (data_saved_folder
                         + "fig_for_basic_neural_analysis_allsessions_basicEvents_neuralGLMfitting_BasisKernelsForContVaris_singlecam/"
                         + cameraID+"/"+savefile_sufix+"/"
                         + recordedanimals[0] + "_neuralGLM/")

        if not os.path.exists(figsavefolder):
            os.makedirs(figsavefolder)

        fig.savefig(figsavefolder + recordedanimals[0]
                    + '_meanFRtraces_PARTNERaligned_byPartnerPC1Encoding_'
                    + savefile_sufix + '_' + cond_toplot_type + '.pdf')
```
